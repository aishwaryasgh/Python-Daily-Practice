### using Object oriented methods


```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
fig = plt.figure()
axis= fig.add_axes([0.1,0.1,0.8,0.8])
axis.plot(x,y)
axis.set_xlabel('x label')
axis.set_ylabel('y label')
axis.set_xlim(min(x), max(x))  # Set x-axis limits
axis.set_ylim(min(y), max(y))  # Set y-axis limit
axis.set_title('Title')
```

```python
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

fig = plt.figure()
axis = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # manual axis setup

axis.plot(x, y, marker='o')  # Add marker for better visibility
axis.set_xlabel('x label')
axis.set_ylabel('y label')
axis.set_title('Title')

plt.show()
```

```python
fig= plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axis2 = fig.add_axes([0.2,0.5,0.4,0.3])

plt.show()
```

```python
fig= plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.8, 1]) 
axis2 = fig.add_axes([0.2,0.5,0.4,0.3])
axis1.plot(x,y)
axis2.plot(y,x)

axis2.set_title('Smaller plot')
axis1.set_title('Larger plot')
plt.show()


```
